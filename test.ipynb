```python
import requests
```

```python
def create_session(filename: str, file_bytes: str):
    payload = {"file": {"filename": filename, "content_type": "application/pdf", "content": file_bytes}}
    response = requests.put("http://localhost:8000/session", json=payload)
    return response.json()

def get_questions(id: str, start_page: int, end_page: int):
    payload = {"session_id": id, "start_page": start_page, "end_page": end_page}
    response = requests.post("http://localhost:8000/questions", json=payload)
    return response.json()
```
